```
import pandas as pd

train_df = pd.read_csv("politeness_train.csv")
dev_df   = pd.read_csv("politeness_val.csv")

print("Train shape:", train_df.shape)
print("Dev shape:", dev_df.shape)


print("\nSample train rows:")
display(train_df.head())
```

```
import re

def clean_spaces(text):
    text = text.strip()
    text = re.sub(r"\s+", " ", text)
    return text

train_df["Sentence"] = train_df["Sentence"].astype(str).apply(clean_spaces)
dev_df["Sentence"]   = dev_df["Sentence"].astype(str).apply(clean_spaces)

train_df.head()
```
