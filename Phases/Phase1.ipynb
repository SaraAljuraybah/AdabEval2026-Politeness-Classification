```
import pandas as pd

train_df = pd.read_csv("politeness_train.csv")
dev_df   = pd.read_csv("politeness_val.csv")

print("Train shape:", train_df.shape)
print("Dev shape:", dev_df.shape)


print("\nSample train rows:")
display(train_df.head())
```

```
import re

def clean_text(text):
    text = text.strip()
    # remove links (http / https)
    text = re.sub(r"http\S+", "", text)


    # remove mentions (handles @user and @ user)
    text = re.sub(r"@\s*\S+", "", text)

    # remove standalone @ left behind (e.g., " @ " )
    text = re.sub(r"\s@\s", " ", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text)

    return text

train_df["Sentence"] = train_df["Sentence"].astype(str).apply(clean_text)
dev_df["Sentence"]   = dev_df["Sentence"].astype(str).apply(clean_text)

train_df.head()

```
